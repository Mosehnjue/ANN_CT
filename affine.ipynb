```python

import numpy as np
import matplotlib.pyplot as plt
import math

import pandas as pd
```

```python
inputs = pd.read_csv('data/affine_1.csv',delimiter=',', header=None) #loading inputs from a different file 
print (inputs.shape)
targets = pd.read_csv('data/UTM1.csv',delimiter=',', header=None) # loading targets
print(targets.shape)

trains = pd.read_csv('data/Cass_f1.csv',delimiter=',', header=None) #loading inputs from a different file 
print (trains.shape)
trains_labels = pd.read_csv('data/utm_test1.csv',delimiter=',', header=None) # loading targets
print (trains_labels.shape)
test = pd.read_csv('data/Xtest.csv',delimiter=',', header=None) # loading targets
print(test.shape)


labels = pd.read_csv('data/yTest.csv',delimiter=',', header=None) # loading targets
print(labels.shape)
```

```python
test
```

```python
labels
```

```python

affine= inputs.to_numpy()
EnUTM = targets.to_numpy()
print (type(affine))
print (type(inputs))
print (type(EnUTM))
print (EnUTM.shape)
```

```python
A = affine
print(A)
At=A.T
```

```python
ATA = np.dot(At,A)
ATA
L = EnUTM
L.shape
```

```python
ATL= np.dot(At,L)
ATL.shape
```

```python
from numpy.linalg import inv
m_inv=inv(ATA)*-1
m_inv
```

```python
x1=-(np.dot(m_inv,ATL)) 
print(x1)
a =x1[0]
b = x1[1]
c=x1[2]
d = x1[3]
e =x1[4]
f = x1[5]
```

```python
ff = trains.to_numpy()
ff
```

```python
ff = trains.to_numpy()
f1 = ff[:,0]*a

f2 = ff[:,1]*b

f3 =  ff[:,0]*d

f4 =  ff[:,1]*e
```

```python
east1 = c+ f1 + f2
east1
```

```python
north1 = f +f3+f4

north1
```

```python
ene_1 = pd.DataFrame(east1)
ene_1.shape
#type(ene1)
```

```python
ene_2 = pd.DataFrame(north1)
ene_1.shape
```

```python
m1 =ene_1.join(ene_2, on= None, how ='right', lsuffix = '_left', rsuffix = '_right')
m1
```

```python
m1.to_csv('yy.csv', header=False, index=False)
```

```python
ene1 = pd.read_csv('yy.csv',delimiter=',', header=None, ) # loading targets
print(ene1.shape)
```

```python
enen1 =ene1.to_numpy()
```

```python
error1 = trains_labels - ene1
error1.to_csv('yy.csv', header=False, index=False)
```

```python
error1
```

```python
import math
from sklearn.metrics import mean_squared_error
mse_true  = mean_squared_error(trains_labels, ene1)
mse_true
```

```python
rmse = math.sqrt(mse_true)
rmse
```
